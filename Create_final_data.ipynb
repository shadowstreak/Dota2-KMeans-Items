```python
'''
Code written by Allen Lauer
Goal of this is to pull team matches from the OpenDota database.
Using all of the assembled data, I can now generate a table that consists of the player info, 
and tabulate all of their decisions in the game. (in this case the 120 or so items you can decide to buy within the game).
Also will hold their performance metrics such as income, and # of wards placed (this is essentially a reconnaissance item that 
players will not hold and will most often not be counted in the 9 end game items.)

I know this is a lot but each player can only purchase up to 9 items at the end of the game, so there are defined play styles
within each match due to the importance of choosing corectly.

For the metrics I will get an average of gold, and an average # of wards placed.

For decades within the game of dota there has been 5 agreed upon defined play styles.

then I am now querying each match ID I gathered before and getting their matches.
'''
```

```python
import pandas as pd
import requests
import json
player_df = pd.read_csv('active_players.csv')
match_df = pd.read_csv('active_matches.csv')
```

```python
import numpy as np
a=np.arange(0,280)

columns = ['team_id','player_ID', 'name', 'total_matches', 'gold', 'wards']
df = pd.DataFrame(columns=columns)
count = 0

print(df)
print(a)
```

```python
for index in a:
    df['item' + str(index)] = 0
```

```python
df
```

```python
df['team_id'] = player_df['team_id']
df['player_ID'] = player_df['player_ID']
df['name'] = player_df['player_name']
```

```python
df = df.fillna(0)
```

```python
for row in match_df.matches:
    with open('matches/archive/' + str(row) + '.json', 'r') as json_file:
        data = json.load(json_file)
        for player_data in data['players']:
            #print(player_data['account_id'])
            try:
                x = df.set_index('player_ID').index.get_loc(player_data['account_id'])
                #print('match', d["players"][i]["account_id"], i, x)

                gold = df.loc[x, 'gold']
                gold2 = gold + player_data["gold_per_min"]
                df = df.set_value( x , 'gold', gold2)

                wards = df.loc[x, 'wards']
                sumwards = wards + player_data["observer_uses"]
                df = df.set_value( x , 'wards', sumwards)
                
                numMatch = df.loc[x, 'total_matches']
                sumMatch = numMatch + 1
                df = df.set_value( x , 'total_matches', sumMatch)
                try:
                    for backpack in range(3):
                        index_find = player_data['item_'+str(backpack)]                            
                        itemadd = df.loc[x, 'item' + str(index_find)]
                        sumadd = itemadd + 1
                        df = df.set_value( x , 'item' + str(index_find), sumadd)

                    for y in range(6):
                        z = player_data['item_'+str(y)]
                        itemadd = df.loc[x, 'item' + str(z)]
                        sumadd = itemadd + 1
                        df = df.set_value( x , 'item' + str(z), sumadd)
                except:
                    print('not enough item slots')
            except:
                print('error in locating user')
```

```python
df = df.replace(0,np.nan).dropna(axis=1,how="all")
```

```python
df = df.replace(np.nan,0)
```

```python
df['avg_gold'] = (df.gold/df.total_matches)
df['avg_wards'] = (df.wards/df.total_matches)
```
