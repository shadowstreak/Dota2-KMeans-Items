```python
import pandas as pd
df = pd.read_csv('teams.csv')
df.shape
```

```python
df
```

```python
import json
import numpy as np
id_list = list()
name_list = list()
team_list = list()

playercount = 1
for row in df.ID:
    with open('teams/team_players_' + str(row) + '.json') as json_data:
        d = json.load(json_data)
        for record in d:
            print(record['name'], record['is_current_team_member'])
            print()
            if record['is_current_team_member'] == True:
                print('inside')
                id_list.append(record['account_id'])
                name_list.append(record['name'])
                team_list.append(row)

x = {'player_ID': id_list, 'player_name': name_list, 'team_id': team_list}
player_df = pd.DataFrame(data=x)
```

```python
player_df.shape
```

```python
player_df
```

```python
totalList = df.ID.unique()
activeList = player_df.team_id.unique()
removeList = set(totalList)^set(activeList)
removeList
```

```python
for index in removeList:
    unique_index = pd.Index(df.ID)
    row_index = unique_index.get_loc(index)
    df = df.drop(df.index[row_index])
```
