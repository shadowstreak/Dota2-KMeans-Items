```python
import pandas as pd
import requests
import json
df = pd.read_csv('active_matches.csv')
df.shape
```

```python
df
```

```python
for row in df.matches:
    r = requests.get('https://api.opendota.com/api/matches/'+str(row)')
    team = r.json()
    with open('matches/archive/' + str(row) + '.json', 'w') as outfile:
        json.dump(team, outfile)
    

    
```
