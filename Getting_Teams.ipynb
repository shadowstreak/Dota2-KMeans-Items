```python
import json
import numpy as np
import pandas as pd
id_list = list()
team_list = list()

with open('master_team_list.json') as json_data:
    d = json.load(json_data)
    cap = 0
    for record in d:
        if cap < 50:
            id_list.append(record['team_id'])
            team_list.append(record['name'])
            cap = cap + 1
        else:
            break


d = {'team': team_list, 'ID': id_list}
team_df = pd.DataFrame(data=d)
```
